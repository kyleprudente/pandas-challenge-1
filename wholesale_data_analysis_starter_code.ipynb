## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['Subtotal'] = df['unit_price'] * df['qty']
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

def shipping_price(x):
    if x > 50:
        return 7*x
    return 10*x

df['Total_weight'] = df['qty'] * df['unit_weight']
df['Shipping_Price'] = df['Total_weight'].apply(shipping_price)
df.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df['total_price'] = (df['Subtotal'] + df['Shipping_Price']) * 1.0925
df.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['total_cost'] = (df['unit_cost']*df['qty']) + df['Shipping_Price']
df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
df['Profit']= df['total_price']-df['total_cost']
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
sorted_df = df.sort_values(by='qty', ascending=False)
top_client_ids = sorted_df['client_id'][:5]
for client_id in top_client_ids:
    client_data = sorted_df.loc[sorted_df['client_id'] == client_id]
    total_qty = client_data['qty'].sum()
    total_spending = client_data['total_price'].sum()
    print(f"Client ID: {client_id}")
    print(f"Total Quantity: {total_qty:,.2f}")
    print(f"Total Spending: ${total_spending:,.2f}")
    print("-" * 30)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
summary_data = []
for client_id in top_client_ids:
    client_data = sorted_df.loc[sorted_df['client_id'] == client_id]
    total_units = client_data['qty'].sum()
    total_shipping = client_data['Shipping_Price'].sum()
    total_revenue = client_data['total_price'].sum()
    total_profit = client_data['Profit'].sum()
    summary_data.append([client_id, total_units, total_shipping, total_revenue, total_profit])
summary_df = pd.DataFrame(summary_data, columns=['Client ID', 'Total Units', 'Total Shipping', 'Total Revenue', 'Total Profit'])
summary_df = summary_df.sort_values(by='Total Profit', ascending=False)
summary_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
summary_df.columns = ['Client', 'Units Purchased', 'Shipping Cost', 'Revenue', 'Profit']
million_formatter = lambda x: f"${x / 1000000:,.2f}M"
summary_df['Shipping Cost'] = summary_df['Shipping Cost'].apply(million_formatter)
summary_df['Revenue'] = summary_df['Revenue'].apply(million_formatter)
summary_df['Profit'] = summary_df['Profit'].apply(million_formatter)
summary_df
```
